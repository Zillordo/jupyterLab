```python
%config IPCompleter.greedy=True

from requests_html import HTML
from requests_html import HTMLSession
import pandas as pd
from pandas import ExcelWriter
from IPython.display import display_html, display
import re
```

```python
session = HTMLSession()
web = session.get('http://portal.chmi.cz/files/portal/docs/meteo/ok/uzemni_teploty_cs.html?fbclid=IwAR3sK4c4qFnk526oozVNsx7pTbdt6TmvJxHwsT2k1abTRg4tFu48d0QpqaQ')
hrefs = []

for a in web.html.find('a'):
    hrefs.append(str(a.attrs['href']))
```

```python
URL_PREFIX = 'http://portal.chmi.cz/files/portal/docs/meteo/ok/'
webs_by_year = []

for href in hrefs:
    webs_by_year.append(session.get(URL_PREFIX + href))
```

```python
tables = []
years = []

for web in webs_by_year:
    for table in web.html.find('table'):
        tables.append(table)
    for year in web.html.find('h2'):
        years.append(year.text)
```

```python
for table, year in zip(tables, years):
    p_table = pd.read_html(table.html)[0]
    p_table['year'] = int(re.search(r'\d+', year).group())
    p_table['Měsíc'] = p_table['Měsíc'].astype(int).div(10)
#     writer = ExcelWriter(f'Verunka_data/{year}.xlsx')
#     p_table.to_excel(writer, sheet_name=year)
#     writer.save()
```
