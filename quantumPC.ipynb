```python
%matplotlib inline
# Importing standard Qiskit libraries and configuring account
import qiskit as q
from qiskit import IBMQ
from qiskit.tools.monitor import job_monitor
# Loading your IBM Q account(s)
provider = IBMQ.load_account()
```

```python
circuit = q.QuantumCircuit(2, 2)
circuit.x(0)
circuit.cx(0, 1)
circuit.measure([0, 1], [0, 1])
```

```python
circuit.draw(output="mpl")
```

```python
IBMQ.providers()
```

```python
provider = IBMQ.get_provider("ibm-q")
```

```python
backend = provider.get_backend("ibmq_burlington")
job = q.execute(circuit, backend=backend, shots=1000)
job_monitor(job)
```

```python
from qiskit.visualization import plot_histogram

result = job.result()
counts = result.get_counts(circuit)

plot_histogram([counts], legend=["Device"])
```

```python
circuit_s = q.QuantumCircuit(2, 2)
circuit_s.h(0)
circuit_s.cx(0 ,1)
circuit_s.measure([0, 1], [0, 1])
circuit_s.draw(output="mpl")
```

```python
job_s = q.execute(circuit_s, backend=backend, shots=500)
job_monitor(job_s)
```

```python
result_s = job_s.result()
counts_s = result_s.get_counts(circuit_s)
plot_histogram([counts_s])
```
