```python
import numpy as np
from enum import Enum
import nnfs
from nnfs.datasets import spiral_data

np.random.seed(0)
%config IPCompleter.greedy=True
nnfs.init()
```

```python
X, y = spiral_data(100, 3)
```

```python
class F():
    def relu(x):
        return np.maximum(0, x)
    
    def sigmoid(z):
        return 1/(1 + np.exp(-z))
    
    def soft_max(x):
        exp_values = np.exp(x - np.max(x, axis=1, keepdims=True))
        probabilities = exp_values / np.sum(exp_values, axis=1, keepdims=True)
        return probabilities
    
    def crossentropyLoss(y_pred, y_true):
        samples = len(y_pred)
        y_pred = y_pred[range(samples), y_true]
        neg_log_p = -np.log(y_pred)
        data_loss = np.sum(neg_log_p) / samples
        return data_loss
```

```python
class Layer_Dense:
    def __init__(self, n_inputs: int, n_neurons: int):
        self.weights = 0.10 * np.random.randn(n_inputs, n_neurons)
        self.biases = np.zeros((1, n_neurons))
    
    def linear(self, inputs: np.ndarray) -> np.ndarray:
        return np.dot(inputs, self.weights) + self.biases
```

```python
dense1 = Layer_Dense(2,3)
dense1_out = F.relu(dense1.linear(X))
dense2 = Layer_Dense(3,3)
dense2_out = F.soft_max(dense2.linear(dense1_out))
loss = F.crossentropyLoss(dense2_out, y)
```

```python
predictions = np.argmax(dense2_out, axis=1)
accuracy = np.mean(predictions==y)
print(loss)
print(accuracy)
```
